```python
import torch
```

```python
x_data= [1.0,2.0,3.0]
y_data= [2.0,4.0,6.0]
```

```python
w = torch.Tensor([1.0])
w.requires_grad=True
```

```python
def forward(x):
    return x *w
```

```python
def loss(x,y):
    y_pred = forward(x)
    return (y_pred -y) **2
```

```python
print("predict before training",4,forward(4).item())
```

```python
forward(4)
```

```python
for epoch in range(100):
    for x,y in zip(x_data,y_data):
        l = loss(x,y)
        l.backward()
        print("--grad:",x,y,w.grad.item())
        
        print("---w.grad")
        print(w.grad)

        print("---w.grad.item()")
        print(w.grad.item())

        print("---w.grad.data")
        print(w.grad.data)

        print("---w.data")
        print(w.data)

        print("---w.data -0.01 * w.grad.data")
        print(w.data -0.01 * w.grad.data)
        w.data = w.data - 0.01 * w.grad.data
        
        w.grad.data.zero_()
        
    print("progress:",epoch,l.item())

print("predict after training:",4,forward(4).item())
```
