```python
import numpy as np
import plotly.express as px
import plotly.graph_objs as plotlyObj
```

```python
import pandas as pd
```

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]
```

```python
def forward(x):
    return x * w +b
```

```python
def loss(x,y):
    y_pred = forward(y)
    return (y_pred -y) ** 2
```

```python
w_list = []
b_list = []
for i in np.arange(0.0,4.1,0.1):
    w_list.append(i)
    b_list.append(i)
```

```python
mse_list = np.zeros((len(np.arange(0.0,4.1,0.1)),len(np.arange(0.0,4.1,0.1))))
for w in np.arange(0.0,4.1,0.1):
    for b in np.arange(0.0,4.1,0.1):
        # print('w=',w)
        # print('b=',b)
        loss_sum = 0
        for x_val,y_val in zip(x_data,y_data):
            y_pred_val = forward(x_val)
            loss_val = loss(x_val,y_val)
            loss_sum += loss_val
            # print("\t",x_val,y_val,y_pred_val,loss_val)
        mse_loss = loss_sum/ len(x_data)
        # print("MSE",mse_loss)
        mse_list[int(10*w)][int(10*b)]=mse_loss
```

```python
mse_list
```

```python
fig = plotlyObj.Figure(
data=plotlyObj.Surface(z=mse_list,x=w_list,y=b_list)
)
fig.update_layout(
    title="权重w和偏置b与 损失MSE的关系",
    width=600,
    height=600,
    scene=dict(
        xaxis_title="x轴 权重w ",
        yaxis_title="y轴 偏置b",
        zaxis_title="z轴 损失MSE", 
    )
)
```
