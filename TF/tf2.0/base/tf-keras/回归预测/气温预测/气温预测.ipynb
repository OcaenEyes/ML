```python
futures = pd.get_dummies(temp_new)
```

```python
labels = futures["actual"]
```

```python
futures = futures.drop(columns=["actual"])
```

```python
# 格式转换
futures_encoding =np.array(futures)
```

```python
futures_encoding
```

```python
from sklearn import preprocessing
```

```python
inpur_futures = preprocessing.StandardScaler().fit_transform(futures_encoding)
```

## 基于Keras构建网络模型
- activation: 激活函数
- kernel_initializer,bias_initializer: 权重与偏置参数的初始化方法，
- kernel_regularizer,bias_regularizer: 要不要加入正则化
- inputs:输入
- units:神经元个数

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(16))
model.add(tf.keras.layers.Dense(32))
model.add(tf.keras.layers.Dense(1))
```

#### compile 相当于对网络进行配置，指定好优化器和损失函数

```python
model.compile(
    optimizer=tf.keras.optimizers.SGD(0.001),
    loss="mean_squared_error"
)
```

```python
model.fit(inpur_futures,labels,validation_split=0.25,epochs=100,batch_size=128)
```

```python
inpur_futures.shape
```

```python
model.summary()
```

#### 更改初始化方法

```python
help(tf.keras.layers.Dense)
```

```python
model1= tf.keras.Sequential()
model1.add(tf.keras.layers.Dense(16,kernel_initializer="random_normal"))
model1.add(tf.keras.layers.Dense(32,kernel_initializer="random_normal"))
model1.add(tf.keras.layers.Dense(1,kernel_initializer="random_normal"))
```

```python
model1.compile(optimizer=tf.keras.optimizers.SGD(0.001),
              loss="mean_squared_error")
```

```python
model1.fit(inpur_futures,labels,validation_split=0.25,epochs=100,batch_size=128)
```

```python
model1.summary()
```

#### 加入正则化惩罚

```python
model2= tf.keras.Sequential()
model2.add(tf.keras.layers.Dense(16,kernel_initializer="random_normal",kernel_regularizer=tf.keras.regularizers.l2(0.03)))
model2.add(tf.keras.layers.Dense(32,kernel_initializer="random_normal",kernel_regularizer=tf.keras.regularizers.l2(0.03)))
model2.add(tf.keras.layers.Dense(1,kernel_initializer="random_normal",kernel_regularizer=tf.keras.regularizers.l2(0.03)))
```

```python
model2.compile(
    optimizer=tf.keras.optimizers.SGD(0.001),
    loss="mean_squared_error"
)
```

```python
model2.fit(inpur_futures,labels,validation_split=0.25,epochs=100,batch_size=64)
```
