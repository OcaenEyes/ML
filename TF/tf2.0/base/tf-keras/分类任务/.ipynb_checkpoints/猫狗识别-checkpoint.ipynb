```python
import os
import tensorflow as tf 
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontManager,FontProperties
import seaborn as sns
```

```python
myfont = FontProperties(fname="../../../../../simhei.ttf")
```

### 数据清洗

```python
base_dir = "../datasets/dogscats/"
```

```python
train_dir = base_dir+"train"
val_dir = base_dir+"val"

train_dog = train_dir+"/dog"
train_cat = train_dir+"/cat"

val_dog = val_dir+"/dog"
val_cat = val_dir+"/cat"

test_dir =  base_dir+"test"
```

- 读取的数据需转换成 tensor(float32)，分别准备训练和验证
- 图像归一到(0-1)之间

```python
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
val_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
```

```python
# 如果one-hot就是categorical， 二分类用 binary就可以
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(64,64),
    batch_size=20,
    class_mode="binary"
)
```

```python
# 如果one-hot就是categorical， 二分类用 binary就可以
validation_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(64,64),
    batch_size=20,
    class_mode="binary"
)
```

```python
validation_generator.class_indices
```

```python
# 如果one-hot就是categorical， 二分类用 binary就可以
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(64,64),
    # batch_size=20,
    # class_mode="binary"
)
```

### 构建模型

```python
model = tf.keras.Sequential()
# conv1d 处理向量， conv2d 处理图形 ， conv3d处理视频
# 
# filters 输出多个特征图
# kernel_size :卷积核大小
# strides: 步长
# padding:填充
model.add(tf.keras.layers.Conv2D(32,(3,3),activation="relu",input_shape=(64,64,3)))
model.add(tf.keras.layers.MaxPool2D(2,2))


model.add(tf.keras.layers.Conv2D(64,(3,3),activation="relu"))
model.add(tf.keras.layers.MaxPool2D(2,2))

model.add(tf.keras.layers.Conv2D(128,(3,3),activation="relu"))
model.add(tf.keras.layers.MaxPool2D(2,2))

# 为全连接层做准备
model.add(tf.keras.layers.Flatten())

model.add(tf.keras.layers.Dense(512,activation="relu"))

# 二分类
model.add(tf.keras.layers.Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(
    loss= tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    metrics=["acc"]
)
```

### 训练网络模型

- 直接.fit是将数据读入内存；  fit_generator是相当于生成器，动态生产所需的batch数据；  最新版本 .fit也具备了生成 generators

- steps_per_epoch相当于一个停止条件，因为生成器会不断的产生batch数据，（控制一个epoch里执行多少个step）

```python
history = model.fit(
    train_generator,
    steps_per_epoch=100, ## images 数量=  batch_size * steps
    epochs=10,
    validation_data=validation_generator,
    validation_steps=50, 
    verbose=2
)
```

```python
history.history
```

```python
import pandas as pd
```

```python
res= pd.DataFrame(history.history)
```

```python
res
```

```python
sns.scatterplot(data=res,x=res.index,y="loss",label="loss")
sns.lineplot(data=res,x=res.index,y="val_loss",label="val_loss")
plt.xlabel("epochs")
plt.ylabel("损失",font=myfont)
```

```python
sns.scatterplot(data=res,x=res.index,y="acc",label="acc")
sns.lineplot(data=res,x=res.index,y="val_acc",label="val_acc")
plt.xlabel("epochs")
plt.ylabel("准确率",font=myfont)
plt.show()
```
