```python
import pandas as pd
import tensorflow as tf 
import plotly.graph_objs as pltgo
import numpy as np
```

```python
data = pd.read_csv("./datasets/credit-a.csv",header=None)
```

```python
data
```

```python
data.iloc[:,-1].value_counts()
```

```python
x = data.iloc[:,:-1]
```

```python
y = data.iloc[:,-1].replace(-1,0)
```

```python
y
```

```python
x
```

```python
model = tf.keras.Sequential()
```

```python
# tf.keras.layers.Dense(4,input_shape=(15,),activation='relu')（隐藏单元数，输入形状-x有多少特征,激活函数）
## param1: 隐藏单元数
## param2: 输入形状
## param3: 激活函数
##### 第一层
model.add(tf.keras.layers.Dense(4,input_shape=(15,),activation='relu'))
##### 第二层
model.add(tf.keras.layers.Dense(4,activation='relu'))
##### 第三层
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['acc'])
```

```python
history = model.fit(x,y,epochs=100)
```

```python
history.history.keys()
```

```python
pltgo.Figure( data=pltgo.Line(x=history.epoch,y=history.history.get('loss')))
```

```python
pltgo.Figure( data=pltgo.Line(x=history.epoch,y=history.history.get('acc')))
```
