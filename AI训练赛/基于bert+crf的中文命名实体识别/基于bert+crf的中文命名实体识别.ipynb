## 基于bert+crf的中文命名实体识别

- **任务描述**：命名实体识别是指识别文本中具有特定意义的实体，主要包括人名、地名、机构名、专有名词等。
    - 实现一个简单的命名实体识别方法，该方法通过BERT+CRF模型预测出文本中文字所对应的标签，再根据标签提取出文本中的实体。
- **方法概述**：
    - 从数据文件中加载数据并进行预处理、构建模型、训练模型、评估模型和测试模型
    - BERT的全称为Bidirectional Encoder Representation from Transformers，是一个预训练的语言表征模型。
        - 在命名实体识别的初级案例中，教程所使用的字符向量为随机生成，相比通过预训练模型生成的字符向量效果较差。
        - 此外，BERT+CRF做命名实体识别效果已经很好了，再加上BiLSTM会导致效率降低，因此本教程模型未使用BiLTSM。
