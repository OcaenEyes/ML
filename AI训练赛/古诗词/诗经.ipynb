```python
data["content"] = data["content"].map(lambda x:"".join(x))
```

```python
shijing_content = "".join(data["content"].tolist())
```

```python
import jieba

from collections import Counter
```

```python
shijing_content_count = Counter(list(jieba.cut(shijing_content,cut_all=True))).most_common()
```

```python
temp_dict = {}
for k,v in shijing_content_count:
    if len(k) >=2:
        temp_dict[k] = v
```

```python
shijing_content_count_dayu2 = [i for i in temp_dict.items()]
```

```python
shijing_content_count_dayu2_wordcloud = pyecharts.charts.WordCloud()
shijing_content_count_dayu2_wordcloud.add("",shijing_content_count_dayu2,word_size_range=[10,66])
```

```python
shijing_content_count_dayu2_wordcloud.load_javascript()
```

```python
shijing_content_count_dayu2_wordcloud.render_notebook()
```
