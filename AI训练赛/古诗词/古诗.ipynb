```python
json_data=[]
for i in shici_file_path:
    with open(os.path.join("./data/chinese-poetry/json",i),encoding='utf-8') as f:
# with open('./data/chinese-poetry/json/poet.song.0.json',encoding='utf-8') as f:
        tmp = json.load(f)
        for index,temp in enumerate (tmp):
            poem=''.join(temp['paragraphs'])
            if (len(poem)>=16) & (len(poem) <=64):
                json_data.append(poem)
```
