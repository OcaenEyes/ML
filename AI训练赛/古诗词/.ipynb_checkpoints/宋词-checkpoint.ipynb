```python
data = pd.DataFrame()
for index,i in enumerate(songci_json_files):
    print(index)
    print(i)
    data = pd.concat([data,pd.read_json(os.path.join('./data/chinese-poetry/ci/',i))])
```

```python
data
```

```python
author_count = data['author'].value_counts().reset_index().rename(columns={'index':'author','author':'count'})
```

```python
rhythmic_count =data['rhythmic'].value_counts().reset_index().rename(columns={'index':'rhythmic','rhythmic':'count'})
```

```python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
import pyecharts

import pyecharts.options as opts
```

```python
author_wordcloud = pyecharts.charts.WordCloud()
```

```python
rhythmic_wordcloud = pyecharts.charts.WordCloud()
```

```python
author_count_tuple = []
for i,row in author_count.iterrows():
    author_count_tuple.append(tuple(row))
    
rhythmic_count_tuple = []
for i,row in rhythmic_count.iterrows():
    rhythmic_count_tuple.append(tuple(row))
```

```python
author_count_tuple
```

```python
author_wordcloud.add("",data_pair=author_count_tuple,word_size_range=[10,160])

author_wordcloud.load_javascript()
```

```python
author_wordcloud.render_notebook()
```

```python
rhythmic_wordcloud.add("",data_pair=rhythmic_count_tuple,word_size_range=[10,160])

rhythmic_wordcloud.load_javascript()
```

```python
rhythmic_wordcloud.render_notebook()
```
