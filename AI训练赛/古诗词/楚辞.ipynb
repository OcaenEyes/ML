```python
data["content"] = data["content"].map(lambda x : "".join(x))
```

```python
chuci_content = "".join(data["content"].tolist())
```

```python
import jieba
import re
import pyecharts
from pyecharts.globals import CurrentConfig,NotebookType
from collections import Counter 
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
chuci_content = "".join(re.findall(r'[\u4e00-\u9fa5]',chuci_content))
```

```python
chuci_content_words = list(jieba.cut(chuci_content,cut_all=True))
```

```python
chuci_content_words_count =Counter(chuci_content_words).most_common()
```

```python
for k,v in chuci_content_words_count:
    if len(k) >= 2:
        print(k,v)
```
