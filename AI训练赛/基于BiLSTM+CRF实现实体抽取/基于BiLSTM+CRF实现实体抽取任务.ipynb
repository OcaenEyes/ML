## 基于BiLSTM+CRF实现实体抽取

- **任务描述**:
    - 我们以地名(LOC), 组织名(ORG), 人名(PER)三类实体为例。
    - 与英文实体抽取不同，中文实体抽取通常需要先对文本进行分词然后再预测序列中单词的类别。
    - 这样会面临一个问题，即在分词中造成的错误会影响到NER的结果。
    - 为此，我们使用基于字级别的实体抽取方法缓解因为分词带来的级联误差。
    - **即以句子中的每一个字为单位进行标注，模型训练和预测。对于训练文本采用BIO标注格式，即对于每一字，都可能有7种类别，如下所示。**
        - 类别名 | O | B-LOC | I-LOC | B-ORG | I-ORG | B-PER | I-PER 
        - 例如，给出句子“小明出生在北京。”中的实体，其中“小”字属于人名的开头词，输出“B-PER”类别；“明”字属于人名的中间词，输出“I-PER”类别；“北”属于地名的开头词，输出“B-LOC”类别。
        - 我们使用准确率（Precision），召回率（Recall）和F1值来衡量模型的效果:
        
        ``` 
            TP: 算法能正确识别的实体
            FP: 算法能识别出错误实体
            FN: 未被算法识别的实体
            Precision: Precision = TP/(TP+FP)
            Recall: Recall = TP/(TP+FN)
            F1: F1 = 2*Precision*Recall/(Precision+Recall)
        ```
